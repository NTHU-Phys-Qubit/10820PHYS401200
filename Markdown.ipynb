# Markdown Cheatsheet

Markdown cheatsheet

## Headers


    # H1 First Level Header
    ## H2 Second Level Header
    ### H3 Third Level Header
    #### H4 Forth Level Header

# H1 First Level Header
## H2 Second Level Header
### H3 Third Level Header
#### H4 Forth Level Header
##### H5 Fifth Level Header
###### H6 Sixth Level Header


## Emphasis

    **Bold text** or __Bold text__
    *Italic text** or _Italic text_

**Bold text** or __Bold text__
*Italic text** or _Italic text_
 
## Bulleted list

    * a
    * b
    * c
    or

    + c

    - a
    - b
    - c
